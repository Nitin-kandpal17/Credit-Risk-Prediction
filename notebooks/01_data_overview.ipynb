# CREDIT RISK PREDICTION - DAY 1: Problem Understanding

### 1. Project Overview
**Objective:** Predict whether a loan applicant is likely to default using credit-related features.

**Dataset:** LendingClub Loan Data or Kaggle‚Äôs Credit Risk Dataset

**Target Variable:** Loan Status (1 = Default, 0 = Paid)

**Tasks:**
- ‚úî Load dataset
- ‚úî Understand structure
- ‚úî Identify target and useful features
- ‚úî Summarize dataset for reporting

```python
# 2. Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# 3. Load Dataset
df = pd.read_csv('../data/credit_risk_dataset.csv')
```

```python
# 4. Basic Overview
print("üîç Dataset Shape:", df.shape)
print("\nüîé First 5 Rows:")
display(df.head())

print("\nüßæ Data Types:")
print(df.dtypes)

print("\nüìä Null Value Count:")
print(df.isnull().sum().sort_values(ascending=False).head(10))
```

```python
# 5. Target Variable Check
# Let's assume 'loan_status' is the target
if 'loan_status' in df.columns:
    print("\n‚úÖ Target column ('loan_status') found.")
    print(df['loan_status'].value_counts())
    sns.countplot(data=df, x='loan_status')
    plt.title("Loan Status Distribution")
    plt.show()
else:
    print("‚ùå 'loan_status' column not found. Please verify column name.")
```

```python
# 6. Basic Stats Summary
print("\nüìà Statistical Summary:")
display(df.describe().T)
```

```python
# 7. Save Dataset Summary
summary = {
    "Total Rows": df.shape[0],
    "Total Columns": df.shape[1],
    "Target Column": "loan_status",
    "Missing Values": df.isnull().sum().sum(),
    "Numerical Columns": df.select_dtypes(include=np.number).shape[1],
    "Categorical Columns": df.select_dtypes(include='object').shape[1]
}

summary_df = pd.DataFrame(list(summary.items()), columns=["Metric", "Value"])
summary_df.to_csv("../reports/dataset_summary.csv", index=False)

print("\n‚úÖ Dataset summary saved to 'reports/dataset_summary.csv'")
```
